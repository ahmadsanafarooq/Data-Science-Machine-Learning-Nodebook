```python
#load dataset
data=pd.read_csv("/content/Crime_Incidents_in_2024.csv")
```

```python
#top values of data use head
data.head()
```

```python
#statistical implementation of data
data.describe()
```

```python
data.info()
```

```python
#missing values
data.isnull().sum()
```

```python
#fill missing values
data['WARD']=data['WARD'].fillna(data['WARD'].mean())
data['ANC']=data['ANC'].fillna(data['ANC'].mode()[0])
data['DISTRICT']=data['DISTRICT'].fillna(data['DISTRICT'].mode()[0])
data.isnull().sum()
```

```python
#drop missing values
data = data.drop(['PSA','NEIGHBORHOOD_CLUSTER','BLOCK_GROUP','CENSUS_TRACT','VOTING_PRECINCT','BID','START_DATE','END_DATE','OCTO_RECORD_ID'], axis=1)
data.isnull().sum()
```

```python
#handle inconsistense data
data['WARD']=data['WARD'].astype(int)
```
