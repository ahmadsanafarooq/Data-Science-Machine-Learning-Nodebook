```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from sklearn.datasets import load_wine
```

```python
dataset=load_wine()
```

```python
dataset
```

```python
df=pd.DataFrame(dataset.data,columns=dataset.feature_names)
df
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
df.describe()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df['target']=dataset.target
```

```python
df
```

```python
df.target.value_counts().plot(kind='bar')
plt.title("Value Counts of target")
plt.xlabel("Target")
plt.xticks(rotation=0)
plt.ylabel("Count")
plt.show()
```

```python
df.columns
```

```python
sns.scatterplot(x='alcohol',y='malic_acid',data=df,hue='target',palette='Set1',s=80)
plt.title("Alcohal & Malic Acid")
plt.show()
```

**1**= *11.0 to almost 13.9*

**0**= *start from almost 12.8 to nearly 15.0*

**2**= *start from almost 12.3 to  nearly 14.4*
