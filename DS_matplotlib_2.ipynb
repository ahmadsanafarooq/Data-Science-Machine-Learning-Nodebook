```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
weather=pd.read_csv('https://raw.githubusercontent.com/alanjones2/dataviz/master/london2018.csv')
```

```python
weather
```

***LINE PLOT***

```python
weather.plot(x="Month",y="Sun")
```

```python
weather.plot(x="Month",y="Rain")
```

```python
weather.plot(x='Month',y='Tmax')
```

```python
weather.plot(x='Month',y='Tmin')
```

***BAR PLOT***

```python
weather.plot.bar(x='Month',y='Rain')
plt.title("Sun Value vs month of the year 2018")
plt.xlabel('Month of the year')
plt.ylabel('Values of the sun')
plt.show()
```

```python
weather.plot.bar(x='Month',y='Rain')
plt.title("Rain vs month of the year 2018")
plt.xlabel('Month of the year')
plt.ylabel('Values of the Rain')
plt.show()
```

```python
weather.plot.bar(x='Month',y='Tmin')
plt.title("Rain vs month of the year 2018")
plt.xlabel('Month of the year')
plt.ylabel('Values of the Rain')
plt.show()
```

```python
weather.plot.bar(x='Month',y='Tmax')
plt.title("Rain vs month of the year 2018")
plt.xlabel('Month of the year')
plt.ylabel('Values of the Rain')
plt.show()
```

```python
weather.plot.bar(x='Month',y=['Sun','Rain','Tmax','Tmin'])
```

```python
weather.plot(x='Month',y=['Sun','Rain','Tmax','Tmin'])
```

***Use of colors in plot***

```python
weather.plot.bar(x='Month',y=["Sun","Rain"],color=["red",'black'])
```

```python
weather.plot(x='Month',y=["Sun","Rain"],color=["red",'black'])
```

```python
weather.plot.bar(x='Month',y=['Sun','Rain','Tmax','Tmin'],color=['red','black','blue','orange'])
```

```python
weather.plot(x="Month", y= ["Sun"],color=["Red"], linestyle=":" ,linewidth=4 ,markerfacecolor="black" ,markersize="15" ,marker="*")
plt.xlabel("Month of the year 2018 (London City)")
plt.ylabel("Values of the rain")
plt.title("Rain value vs Months of the year 2018")
plt.show()
```

```python
weather.plot.bar(x="Month", y= ["Sun", "Rain", "Tmax"],color=["Pink", "olive", "blue"], layout = (2,2), subplots= True)
```

***PIE PLOTS***

```python
plt.figure(figsize = (32,32))
weather.plot.pie(subplots=True, layout=(2,3))
```
