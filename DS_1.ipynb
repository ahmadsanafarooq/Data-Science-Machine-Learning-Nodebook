***Missing Values***

```python
missing_values=df.isnull().sum()
print("Missing values\n\n", missing_values)
```

```python
df.dropna(inplace=True)#drop missing values
```

***Duplicate Data***

```python
duplicate_records=df[df.duplicated()]
print("Duplicate records\n\n",duplicate_records)
```

```python
#total duplicate values
duplicated_sum=df.duplicated().sum()
print("Number of duplicate records",duplicated_sum)
```
